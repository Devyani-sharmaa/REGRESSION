```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.DataFrame({"Age":[21,48,32,41,20,35,20,23,42,34,24,22,23,25,43,44,25,30,31],
                        "Bought Insurence":["no","yes","yes","yes","no","yes","no","no","yes","yes",
                                            "no","no","no","no","yes","yes","no","yes","yes"]})
```

```python
dataset.to_csv("Data.csv",index=False)
```

```python
data=pd.read_csv("Data.csv")
data
```

```python
plt.scatter(data["Age"],data["Bought Insurence"])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
data["Bought Insurence"]=encoder.fit_transform(data["Bought Insurence"])
```

```python
data
```

train test split

```python

x = data.iloc[:, :-1]
y = data["Bought Insurence"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
model.predict([[28]])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```
