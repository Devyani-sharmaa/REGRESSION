```python
import numpy as np
```

```python
x = 6 * np.random.rand(200, 1) - 3
y = 0.8 * x ** 2 + 0.9 * x + 2 + np.random.randn(200, 1)
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2)
```

```python
x_train.shape
```

```python
plt.scatter(x_train,y_train,color="blue",label="training")
plt.scatter(x_test, y_test, color = "red", label = "testing data")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
plt.scatter(x_train, y_train, color = "green", label = "traning data")
plt.scatter(x_test, y_test, color = "red", label = "testing data")
plt.plot(x_test, lr.predict(x_test), color = "blue", label = "Predicted line")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, lr.predict(x_test))
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree = 2)
```

```python
x_train
```

```python
x_train_new=poly.fit_transform(x_train)
```

```python
x_train_new
```

```python
x_test_new=poly.fit_transform(x_test)
```

```python
x_test_new
```

```python
my_model=LinearRegression()
```

```python
my_model.fit(x_train_new,y_train)
```

```python
my_model.predict(poly.fit_transform([[2.92978729]]))
```

```python
plt.scatter(x_train, y_train, color = "green", label = "traning data")
plt.scatter(x_test, y_test, color = "red", label = "testing data")
plt.plot(x_test, my_model.predict(poly.fit_transform(x_test)), color = "blue", label = "Predicted line")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
```

```python
poly.fit_transform([[3]])
```
